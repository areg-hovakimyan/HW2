```python
from Bandit import EpsilonGreedy, ThompsonSampling, Visualization, comparison
import numpy as np
```

```python
bandit_probabilities = np.array([0.1, 0.3, 0.5, 0.9])

epsilon = 0.1
epsilon_greedy_bandit = EpsilonGreedy(bandit_probabilities, epsilon)

thompson_sampling_bandit = ThompsonSampling(bandit_probabilities)

N = 20000

epsilon_greedy_rewards = epsilon_greedy_bandit.experiment(N)
thompson_sampling_rewards = thompson_sampling_bandit.experiment(N)

viz = Visualization()
viz.plot1([epsilon_greedy_rewards, thompson_sampling_rewards], ['Epsilon-Greedy', 'Thompson Sampling'])
viz.plot2(epsilon_greedy_rewards, thompson_sampling_rewards)

thompson_sampling_bandit.report()
epsilon_greedy_bandit.report()
```

```python
comparison(thompson_sampling_rewards, epsilon_greedy_rewards)
```
